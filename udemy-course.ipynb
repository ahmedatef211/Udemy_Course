```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/udemy-courses-dataset/UdemyCoursesDataset.csv')

values = df['is_paid'].value_counts()
labels = values.index
colors = ['#1f77b4', '#2ca02c']
explode = (0, 0.1)


plt.pie(values, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90, wedgeprops=dict(width=0.3))

plt.title('Paid and Free Courses Distribution (Donut Chart)')
plt.axis('equal')
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/udemy-courses-dataset/UdemyCoursesDataset.csv')

sns.histplot(data=df, x='price', hue='is_paid', multiple='stack', palette=['#2ca02c', '#1f77b4'], bins=30)

plt.title('Price Distribution of Paid and Free Courses')
plt.xlabel('Price ($)')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/udemy-courses-dataset/UdemyCoursesDataset.csv')

df['year'] = pd.to_datetime(df['published_timestamp'], errors='coerce').dt.year

counts = df.groupby(['year', 'is_paid']).size().unstack().fillna(0)

counts.plot(kind='line', marker='o', color=['#2ca02c', '#1f77b4'])

plt.title('Number of Paid and Free Courses per Year')
plt.xlabel('Year')
plt.ylabel('Number of Courses')
plt.legend(title='Course Type')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/udemy-courses-dataset/UdemyCoursesDataset.csv')

top10 = df.nlargest(10, 'num_subscribers')

plt.barh(top10['course_title'], top10['num_subscribers'], color='#1f77b4')
plt.xlabel('Number of Subscribers')
plt.ylabel('Course Title')
plt.title('Top 10 Courses by Number of Subscribers')
plt.gca().invert_yaxis()
plt.show()
```
